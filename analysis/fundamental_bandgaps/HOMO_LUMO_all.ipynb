```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```python
df = pd.read_csv('GA_data_v1234.csv')
df = df.sort_values(by=['PCE'], ascending=True)
df = df.reset_index(drop=True)
df
```

```python
fig, (ax, ax2)  = plt.subplots(1, 2, figsize=(10, 4),constrained_layout=True, gridspec_kw={'width_ratios': [10, 1]})

df_above20 = df[df.PCE > 20]
viridis = cm.get_cmap('inferno_r',len(df_above20))
newcolors = viridis(np.linspace(0, 1, len(df_above20)))

entry_count = len(df)

top_PCE = []
count = 0

for i in range(entry_count):

    if df['PCE'][i] < 20:

        ax.scatter(x=df['fund bg'][i], y=df['LUMO (eV)'][i], c='darkgrey')
        ax.scatter(x=df['fund bg'][i], y=df['HOMO (eV)'][i], c='darkgrey')
        #count +=1
        #top_PCE.append(df['PCE'][i])

    else:
        ax.scatter(x=df['fund bg'][i], y=df['LUMO (eV)'][i], c=[newcolors[count]])
        ax.scatter(x=df['fund bg'][i], y=df['HOMO (eV)'][i], c=[newcolors[count]])
        count +=1
        top_PCE.append(df['PCE'][i])

#for i in range(len(df)-50, len(df)):

    #top_PCE.append(df['PCE'][i])
    #ax.vlines(df['opt bg (nm)'][i], [0], df['PCE'][i], colors = newcolors[count])
    #count +=1

min_PCE = min(top_PCE)
max_PCE = max(top_PCE)

cmap = mpl.cm.inferno_r
norm = mpl.colors.Normalize(vmin = min_PCE, vmax = max_PCE)


colbar = mpl.colorbar.ColorbarBase(ax2, cmap=cmap, norm=norm, orientation='vertical')
colbar.set_label('PCE', fontsize=16)


ax.set_xlabel('Fundamental Bandgap (eV)', fontsize=16)
ax.set_ylabel('Energy (eV)', fontsize=16)
ax.text(x=2.25, y=-8, s= 'HOMO', fontsize=14)
ax.text(x=2.25, y=-1.5, s= 'LUMO', fontsize=14)
#ax.axes.xaxis.set_visible(False)

plt.savefig('homo_lumo_fundbg_v1234.png')
plt.savefig('homo_lumo_fundbg_v1234.pdf', dpi=600)
```
